# 13. Which transactions have a UserID not in users?

# 14. Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join)

```python
#using inner join merging transactions and users dataframe on UserId field to fetch only matching rows with UserId
transactions.merge(users, on='UserID',how='inner')
```

# 15.Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)

```python
#merging transactions and users dataframe using full outer join to show all records with /with out matching userId field in both dataframes
transactions.merge(users,on='UserID',how='outer')
```

```python
users['Registered'] = pd.to_datetime(users.Registered)
users['Cancelled'] = pd.to_datetime(users.Cancelled)
transactions['TransactionDate'] = pd.to_datetime(transactions.TransactionDate)
sessions['SessionDate'] = pd.to_datetime(sessions.SessionDate)
```

# 16. Determine which sessions occurred on the same day each user registered

```python
#merge sessions dataframe with users dataframe using equi join (inner join) from sessions using SessionDate and UserId and on Users using Registered and UserId
sessions.merge(users,how='inner',left_on=['SessionDate','UserID'],right_on=['Registered','UserID'])
```

# 17. Build a dataset with every possible (UserID, ProductID) pair (cross join)

```python
# users.shape[0] will give total number of rows in users and same in products.shape[0]
#np.repeat will generage an array from 1 to number of rows in each dataframe
# after creating two dataframes by appending key values with row numbers
# do merge of both dataframes based on key value and it will result cartesian of UserId and ProductId dataframe
df1 = pd.DataFrame({'key':np.repeat(1,users.shape[0]),'UserID':users.UserID})
df2 = pd.DataFrame({'key':np.repeat(1,products.shape[0]),'ProductID':products.ProductID})
user_products = pd.merge(df1,df2,on='key')[['UserID','ProductID']]
user_products
```

# 18. Determine how much quantity of each product was purchased by each user

```python
#first group by transactions based on UserID and ProductID, then select quantity and do by sum() operation
##In the class it was asked to show Product Name instead of Product Id to work on merging two tables 

#transactions.merge(products,on='ProductID',how='inner').groupby(['UserID','Product'])['Quantity'].sum()
#transactions.groupby(['UserID','ProductID'])['Quantity'].sum()
```

```python
#determine how much quantity of each product was purchased by user
pd.merge(user_products,transactions,how='left',on=['UserID','ProductID']).groupby(['UserID','ProductID']).apply(lambda x:pd.Series(dict(quantity=x.Quantity.sum()))).reset_index().fillna(0)
```
